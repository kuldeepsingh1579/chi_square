```python
import scipy.stats as stats

# Sample data characteristics
sample_mean = 50   
sample_stddev = 5  
sample_size = 30  

# Set the confidence level
confidence_level = 0.95

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (sample_stddev / (sample_size**0.5))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"95% Confidence Interval: {confidence_interval}")
```

```python
import scipy.stats as stats

# Observed frequencies (counts) of each color in the sample
observed_counts = [25, 22, 18, 10, 12, 23]  # These are your actual counts for each color

# Expected frequencies based on the expected distribution
expected_distribution = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]
total_sample_size = sum(observed_counts)
expected_counts = [p * total_sample_size for p in expected_distribution]

# Perform the chi-square goodness-of-fit test
chi_square, p_value = stats.chisquare(observed_counts, expected_counts)
print(chi_square,p_value)

# Define the significance level
alpha = 0.05

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution differs from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")
```

```python
import scipy.stats as stats

# Sample size (population size)
total_population = 500

# Number of individuals who smoke
smokers = 60

# Sample proportion
sample_proportion = smokers / total_population
print("sample_proportion =" ,sample_proportion)

# Set the confidence level
confidence_level = 0.95

# Calculate the standard error
standard_error = (sample_proportion * (1 - sample_proportion) / total_population) ** 0.5
print("standard_error=", standard_error)

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * standard_error

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)

print(f"95% Confidence Interval for Proportion: {confidence_interval}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Generate values for the x-axis
x = np.linspace(0, 30, 400)  # Adjust the range as needed

# Calculate the chi-square probability density function
pdf = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, pdf, label=f"Chi-Square ({df} df)")
plt.fill_between(x, pdf, where=(x >= 15), color='gray', alpha=0.5)

# Label the axes
plt.xlabel("Chi-Square Statistic")
plt.ylabel("Probability Density")

# Title
plt.title(f"Chi-Square Distribution with {df} degrees of freedom")

# Add a vertical line at chi-square statistic 15
plt.axvline(x=15, color='red', linestyle='--', label='Chi-Square = 15')

# Legend
plt.legend()

# Show the plot
plt.show()
```

```python
import scipy.stats as stats

# Sample size
sample_size = 1000

# Number of people who prefer Coke in the sample
prefer_coke = 520

# Sample proportion
sample_proportion = prefer_coke / sample_size

# Set the confidence level
confidence_level = 0.99

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (sample_proportion * (1 - sample_proportion) / sample_size) ** 0.5

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)

print(f"99% Confidence Interval for Proportion: {confidence_interval}")
```
